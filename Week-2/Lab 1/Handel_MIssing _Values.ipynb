## <font color="maroon"><h4 align="center">Handling Missing Data - replace method</font>

```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("weather_data.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

**Replacing single value**

```python
new_df = df.replace([-99999, 0], np.nan)
new_df
```

```python
type(df.loc[3, "event"])
```

```python
new_df = df.replace([-99999, "0"], np.nan)
new_df
```

**Replacing list with single value**

```python
new_df = df.replace(to_replace=[-99999,-88888], value=0)
new_df
```

**Replace Per Column**

```python
new_df = df.replace({
        'temperature': -99999,
        'windspeed': -99999,
        'event': '0'
    }, np.nan)
new_df
```

**Replacing by using mapping**

```python
new_df = df.replace({
        -99999: np.nan,
        '0': 'Sunny',
    })
new_df
```

####List ([]) = Many Values ➜ One Value
####Mapping ({old:new}) = Each Value ➜ Different Value ✅

**Regex**

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True)
new_df
```

**Replace list with another list**

```python
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
df.replace(
    ['poor', 'average', 'good', 'exceptional'],
    [1, 2, 3, 4]
)
```
