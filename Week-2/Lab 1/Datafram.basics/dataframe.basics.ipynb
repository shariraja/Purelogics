# <font color="purple"><h3 align="center">Pandas Concatenate Tutorial</h3></font>

```python
import pandas as pd

pakistani_weather = pd.DataFrame({
    "city": ["karachi","lahore","peshawar"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
pakistani_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
#concatinate us and pakistani weather

df = pd.concat([pakistani_weather,us_weather])
df
```

```python
df = pd.concat([pakistani_weather,us_weather], ignore_index = True )
df
```

```python
df = pd.concat([pakistani_weather,us_weather],keys = ['Pakistan','US'])
df
```

```python
#show city temperate and humidity of us cities

df.loc['US']
```

```python
#show information of pakistani cities (city, temperature, humidity)

df.loc['Pakistan']
```

## <font color='blue'>Concatenation Using Index</font>

```python
temperature_df = pd.DataFrame({
    "city": ["karachi","lahore","peshawar"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["lahore","karachi"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
#concatenate the temperature and windspeed of Lahore, Karachi and Peshawar
df = pd.concat([temperature_df,windspeed_df],axis = 1)
df
```

## <font color="maroon"><h4 align="center">Pandas Group By</font>

```python
import pandas as pd
df = pd.read_csv("weather_by_cities.csv")
df
```

### For this dataset, get following answers,
#### 1. What was the maximum temperature in each of these 3 cities?
#### 2. What was the average windspeed in each of these 3 cities?


```python
temp_max = df.groupby('temperature')['temperature'].max()
temp_max.idxmax()
```

```python
ws_avg = df.groupby('windspeed')['windspeed'].mean()
ws_avg
```

```python
df['windspeed'].mean()
```

```python
g = df.groupby('city')
g
```

```python
for city, data in g:
    print("city:",city)
    print("\n")
    print("data:",data)
```

# <font color="purple"><h3 align="center">Pandas Merge Tutorial</h3></font>

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
#merge df1 and df2 on citydf_new = pd.merge(df1 , df2 , on='city')
df_new 
```

## <font color='blue'>Type Of DataBase Joins</font>

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

## <font color='blue'>indicator flag</font>

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

## <font color='blue'>suffixes</font>

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```
